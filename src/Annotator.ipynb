```python
# Import Libraries
import os
import glob
from PIL import Image
from matplotlib import cbook
from matplotlib import pyplot as plt
```

```python
items = os.listdir('/home/afsharino/Desktop/Redeemer-Bot/dataset')
items = [item.split('-')[0] for item in items]
print(f'{len(items)} captchas are annotated now.')

for image in glob.glob('/home/afsharino/Desktop/Redeemer-Bot/downloaded_images/*'):
    image_name = image.split('/')[-1].split('.')[0]
    
    if image.split('/')[-1].split('.')[0] not in items:

        try:
            with Image.open(image) as img:
                plt.imshow(img)
                plt.title(image_name)
                plt.show()

                captcha_number = input("Enter number: ")

                file_name = image.split('/')[-1].split('.')[0] + '-' + captcha_number + '.png'
                file_path = os.path.join('/home/afsharino/Desktop/Redeemer-Bot/dataset', file_name)
                img.save(file_path, 'png')

        except Exception as e:
            print(e)
            print(f"can't import {image.split('/')[-1]}.")

    else:
        continue
```
